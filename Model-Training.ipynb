# Model Training

```python
from data_book import DataBook
import sklearn as skl
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn import svm
from sklearn.linear_model import LogisticRegression
print(skl.__version__)
raw_data_path = './Data-v1/OutputEnriched1006.xls'
```

## Loading an Excel file as a DataBook object

```python
dBook = DataBook()
dBook.load_file(raw_data_path)
dBook.get_source_data()
```

## Add features

We compute features for each cell, that is, for each cell we process its context

```python
dBook.pre_process_data()
dBook.get_data()
```

## Add negative cases

Negative cases are cells with formula marked as inconsistent, that are actual errors. This is required for training, cause we are missing enough supervision. Note this is adding the *Label* column to the dataframe (*False* means a cell with an inconsistent formula is not an error).

```python
dBook.add_negative_cases('./Data-v1/areas.txt')
df=dBook.get_data(all_columns=False).copy(deep=True)
df.reset_index(inplace=True)
df
```

## Split traint and test data

We extract the name of the Excel sheet, and we filter the dataframe over sheets that contains some positive cases. This is what we are giving to *traintestsplit*.

```python
df['SheetName'] = df.apply(lambda r: r['key'].split('!')[0], axis=1)
df
```

```python
sheet_names = df[df.Label==True]['SheetName'].unique()
sheet_names
```

```python
red_df = df[df['SheetName'].isin(sheet_names)]
```

```python
features = [c for c in df.columns if c not in ['Label', 'key', 'lineageId', 'SheetName']]
X = red_df[features]
Y = red_df['Label']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.2)
```

## Train and model

```python
# LRC = LogisticRegression()
LRC = LogisticRegression(random_state=0, solver='lbfgs', multi_class='ovr')
LRC.fit(X_train, Y_train)

predicted_LRC = LRC.predict(X_test)

# Use Confusion matrix to visualise results
# confusion_matrix(predicted, Y_test)
print(classification_report(Y_test, predicted_LRC))
```

```python
SVM = svm.LinearSVC()
SVM.fit(X_train, Y_train)

predicted_SVM = SVM.predict(X_test)
print(classification_report(Y_test, predicted_SVM))
```
